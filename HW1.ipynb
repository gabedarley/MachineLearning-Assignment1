```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
d = 2

# Generate random training data
X = np.random.uniform(-1, 1, size = (N, d+1))
X[:,0] = 1

#Generate random target function: f(x) = w^Tx
w = np.random.uniform(-1, 1, size = (d + 1))
x = np.linspace(-1, 1, 100)
y = -(w[0]/w[2]) + (-(w[1] * x)/w[2])


#Compute true labels for the training data
Y = np.sign(np.dot(X,w))
ind_pos = np.where(Y==1)[0] #positive examples
ind_neg = np.where(Y==-1)[0] #negative examples

#Plot points
plt.clf()
plt.plot(X[ind_pos, 1], X[ind_pos, 2], 'ro')
plt.plot(X[ind_neg, 1], X[ind_neg, 2], 'bx')
plt.plot(x, y, '-r', label='target function')


#Adjust Perceptron
#Establish boolean and index variables for adjustment
correct_vector = False
error_var = False
i = 0
t = 1

#Generate starting test hypothesis
f = np.random.uniform(-1, 1, size = (d + 1))
# plt.plot(f)

#While loop to run through x values and check against perceptron
while correct_vector == False:
    Z = np.sign(np.dot(X,f))
    
    for i in range(len(X)):
        if Y[i] == 1 and Z[i] == -1:
            break
        elif Y[i] == -1 and Z[i] == 1:
            break
    if i == len(X) - 1:
        correct_vector = True
        break
    else:
        f = f + np.dot(X[i,:], Y[i])
        t += 1
        i = 0

print("Number of updates = %d" % t)

#Plot specs
plt.xlabel('x')
plt.ylabel('y')
plt.title('Problem 1.4')

#Plot the final hypothesis
f_y = -(f[0]/f[2]) + (-(f[1] * x)/f[2])
plt.plot(x, f_y, label='final hypothesis')
plt.legend(loc = 'upper left')

plt.show()
```

b. For the first trial, the perceptron was only updated 5 times before reaching the final hypothesis. While the target function and final hypothesis are similar, they are not necessarily the same. The final hypothesis is only an approximation, still sharing the same ability to separate the two data sets, but likely to be a different function due to the limits of having a finite number of data points.

c. For the second trial, the perceptron was updated 6 times before reaching the final hypothesis. This is a (possibly coincidentally) very similar number to our first trial. Again the final hypothesis does not match up exactly with the target function.

d. When using a data set of 100, the perceptron actually updated itself 61 times. This is obviously a much larger number of updates when compared with our original sample size of 20 points. However, the final hypothesis is much closer to the target function. This improved accuracy is likely due to the increased amount of information (data points) we have.

e. When using a data set of 1000, the perceptron updated itself 202 times. This is again an even larger number than with the last data set, but again we can see a greater correlation between the final hypothesis and target function, to the point where they are virtually indistinguishable.
